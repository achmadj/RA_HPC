```python
import numpy as np
import matplotlib.pyplot as plt
import re
from icecream import ic

with open('/clusterfs/students/achmadjae/RA/03_deep6D/convNd/log/loss.txt', 'r') as f:
    lines = f.readlines()

train_loss = []
"""
each line is in the form:
tensor(0.5998, device='cuda:0')
"""
for line in lines:
    if 'tensor' in line:
        train_loss.append(float(re.findall(r'\d+\.\d+', line)[0]))

plt.plot(train_loss)
plt.yscale('log')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.show()
```
